```python
import graphviz

# Create a directed graph
dot = graphviz.Digraph(comment='Graph with Multi-Character Nodes')

# Add nodes with multiple characters
dot.node('Ax')
dot.node('By')
dot.node('Cz')

# Define edges correctly (unpack each element from the list)
for edge in ['Ax:By', 'By:Cz', 'Cz:Ax']:
  tail, head = edge.split(':')  # Split the edge definition at ':'
  dot.edge(tail, head)  # Add the edge using separate tail and head nodes

# Render and display the graph in the Jupyter Notebook
dot
```
