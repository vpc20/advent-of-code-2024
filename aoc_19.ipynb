```python
def count_arrangements(patterns, designs):
    """
    Counts the total number of ways to arrange towels for each design.

    Args:
        patterns: A list of available towel patterns.
        designs: A list of desired designs.

    Returns:
        The total number of arrangements for all designs.
    """

    dp = [[0] * (len(design) + 1) for design in designs]
    pattern_lengths = [len(pattern) for pattern in patterns]

    for i in range(len(designs)):
        dp[i][0] = 1  # Empty design can be constructed in one way

    for i in range(len(designs)):
        for j in range(1, len(designs[i]) + 1):
            for k in range(len(patterns)):
                if j >= pattern_lengths[k] and designs[i][j - pattern_lengths[k]:j] == patterns[k]:
                    dp[i][j] += dp[i][j - pattern_lengths[k]]

    total_arrangements = 0
    for i in range(len(designs)):
        total_arrangements += dp[i][-1]

    return total_arrangements

# Example usage:
patterns = ['r', 'wr', 'b', 'g', 'bwu', 'rb', 'gb', 'br']
designs = ['brwrr', 'bggr', 'gbbr', 'rrbgbr', 'ubwu', 'bwurrg', 'brgr', 'bbrgwb']

result = count_arrangements(patterns, designs)
print(result)  # Output: 16
```
