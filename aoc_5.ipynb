```python
rules = '''47|53
97|13
97|61
97|47
75|29
61|13
75|53
29|13
97|29
53|29
61|53
97|53
61|29
47|13
75|47
97|75
47|61
75|61
47|29
75|13
53|13'''

seta = set([rule.split('|')[0] for rule in rules.split()])
print(seta)
setb = set([rule.split('|')[1] for rule in rules.split()])
print(setb)
setc = seta.union(setb)
print(setc)

for n in setc:
    print(f"dot.node('{n}')")
```

```python
import graphviz

# Create a directed graph
dot = graphviz.Digraph(comment='Graph with Multi-Character Nodes')

# Add nodes with multiple characters
dot.node('53')
dot.node('47')
dot.node('61')
dot.node('29')
dot.node('13')
dot.node('75')
dot.node('97')

# Define edges correctly (unpack each element from the list)
for rule in rules.split():
    head, tail = rule.split('|')  # Split the edge definition at '|'
    dot.edge(head, tail)  # Add the edge us

# Render and display the graph in the Jupyter Notebook
dot
```
